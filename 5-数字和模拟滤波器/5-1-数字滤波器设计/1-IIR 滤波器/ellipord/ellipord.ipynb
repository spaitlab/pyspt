# 信号处理仿真与应用 - 数字和模拟滤波器 - 数字滤波器设计
## ellipord

### 低通椭圆滤波器设计

```python
import scipy.signal as signal
import matplotlib.pyplot as plt
import numpy as np

# 定义参数
Wp = 40/500
Ws = 150/500
Rp = 3
Rs = 60

# 计算滤波器阶数和新的归一化截止频率
n, Wp = signal.ellipord(Wp, Ws, Rp, Rs)

# 设计切比雪夫II型带通滤波器
b, a = signal.ellip(n, Rp, Rs, Wp, btype='low')

# 计算频率响应
w, h = signal.freqz(b, a, worN=512, fs=1000)

# 绘制频率响应
plt.figure()
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title("n = %d" % n +" Elliptic Lowpass Filter")
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid()
plt.show()

# 绘制相位响应
plt.figure()
plt.plot(w, np.degrees(np.unwrap(np.angle(h))), 'b')
plt.title('Phase')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [degrees]')
plt.grid()
plt.show()

```

### 带通椭圆滤波器设计

```python
import scipy.signal as signal
import matplotlib.pyplot as plt
import numpy as np

# 定义参数
Wp = [60/500, 200/500]
Ws = [50/500, 250/500]
Rp = 3
Rs = 40

# 计算滤波器阶数和新的归一化截止频率
n, Wp = signal.ellipord(Wp, Ws, Rp, Rs)

# 设计切比雪夫II型带通滤波器
b, a = signal.ellip(n, Rp, Rs, Wp, btype='pass')

# 计算频率响应
w, h = signal.freqz(b, a, worN=512, fs=1000)

# 绘制频率响应
plt.figure()
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title("n = %d" % n +" Elliptic Lowpass Filter")
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid()
plt.show()

# 绘制相位响应
plt.figure()
plt.plot(w, np.degrees(np.unwrap(np.angle(h))), 'b')
plt.title('Phase')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [degrees]')
plt.grid()
plt.show()
```
