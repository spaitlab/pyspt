# 信号处理仿真与应用 - 数字和模拟滤波器 - 数字滤波器设计
## cheb2ord

### 切比雪夫II型滤波器设计

```python
import scipy.signal as signal
import matplotlib.pyplot as plt
import numpy as np

# 定义参数
Wp = 40 / 500  # 归一化通带截止频率
Ws = 150 / 500  # 归一化阻带截止频率
Rp = 3         # 通带最大衰减dB
Rs = 60        # 阻带最小衰减dB

# 计算滤波器阶数和新的归一化截止频率
n, Ws = signal.cheb2ord([Wp], [Ws], Rp, Rs)

# 设计切比雪夫II型滤波器
b, a = signal.cheby2(n, Rs, Ws, analog=False)

# 计算频率响应
w, h = signal.freqz(b, a, worN=512,fs=1000)

# 绘制频率响应
plt.figure(1)
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title('n = 4 Chebyshev Type II Lowpass Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid()
plt.show()

# 绘制相位响应
plt.figure(2)
plt.plot(w, np.degrees(np.unwrap(np.angle(h))), 'b')
plt.title('Phase Response of Chebyshev Type II Lowpass Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [degrees]')
plt.grid()
plt.show()
```

### 切比雪夫II型带通滤波器设计

```python
import scipy.signal as signal
import matplotlib.pyplot as plt
import numpy as np

# 定义参数
Wp = [60 / 500, 200 / 500]  # 归一化通带截止频率
Ws = [50 / 500, 250 / 500]  # 归一化阻带截止频率
Rp = 3                      # 通带最大衰减dB
Rs = 40                     # 阻带最小衰减dB

# 计算滤波器阶数和新的归一化截止频率
n, Ws = signal.cheb2ord(Wp, Ws, Rp, Rs)

# 设计切比雪夫II型带通滤波器
b, a = signal.cheby2(n, Rs, Ws, btype='bandpass', analog=False)

# 计算频率响应
w, h = signal.freqz(b, a, worN=512, fs=1000)

# 绘制频率响应
plt.figure(3)
plt.plot(w, 20 * np.log10(abs(h)), 'b')
plt.title('n = 7 Chebyshev Type II Bandpass Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude [dB]')
plt.grid()
plt.show()

# 绘制相位响应
plt.figure(4)
plt.plot(w, np.degrees(np.unwrap(np.angle(h))), 'b')
plt.title('Phase Response of Chebyshev Type II Bandpass Filter')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Phase [degrees]')
plt.grid()
plt.show()
```
