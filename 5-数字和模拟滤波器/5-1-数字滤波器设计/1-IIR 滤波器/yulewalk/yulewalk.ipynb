```python
import numpy as np
from scipy.signal import freqz
import matplotlib.pyplot as plt
from yulewalker import yulewalk

# 定义参数
order = 8  # 滤波器阶数
f = np.array([0, 0.6, 0.6, 1])  # 频率点
m = np.array([1, 1, 0, 0])  # 阻抗


# 设计Yule-Walker滤波器
b, a = yulewalk(order, f, m)

# 计算频率响应
w, h = freqz(b, a, worN=128)

# 绘制频率响应
plt.figure()
plt.plot(w / np.pi, -20 * np.log10(abs(h)))
plt.title('Yule-Walker Filter Frequency Response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid()
# 在x轴的0.6dB处添加竖线
plt.axvline(x=0.6, color='r', linestyle='--')

plt.show()
```

```python
f1 = np.array([0, 0.55, 0.6, 0.65, 1])
m1 = np.array([1, 1, 0.5, 0, 0])
# 设计Yule-Walker滤波器
b1, a1 = yulewalk(order, f1, m1)

# 计算频率响应
w1, h1 = freqz(b1, a1, worN=128)

# 绘制频率响应
plt.figure()
plt.plot(w / np.pi, -20 * np.log10(abs(h)))
plt.plot(w1 / np.pi, -20 * np.log10(abs(h1)))
plt.title('Yule-Walker Filter Frequency Response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid()
# 在x轴的0.6dB处添加竖线
plt.axvline(x=0.6, color='r', linestyle='--')

plt.show()
```
