## firtype


#### 线性相位滤波器的类型

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from firtype import firtype
 

# 定义fir1函数
def fir1(numtaps, cutoff):
    return signal.firwin(numtaps+1, cutoff)

# 第一个案例
plt.subplot(2, 1, 1)
b1 = fir1(8, 0.5)
print(b1)
plt.stem(b1)
plt.title('Type ' + str(firtype(b1)))

# 第二个案例
plt.subplot(2, 1, 2)
b2 = fir1(9, 0.5)
print(b2)
plt.stem(b2)
plt.title('Type ' + str(firtype(b2)))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal


# Design and plot equiripple Hilbert transformer with even order (8 taps)
filter_order_even = 9
bands_even = [0, 0.2, 0.8, 1]
desired_even = [1, 1]
weights_even = [1, 1]
b1 = signal.remez(filter_order_even, bands_even, desired_even, weights_even, fs=2, type='hilbert')

print(b1)
plt.subplot(2, 1, 1)
plt.stem(b1)
plt.title('Type ' + str(firtype(b1)))

# Design and plot equiripple Hilbert transformer with odd order (9 taps)
filter_order_odd = 10
bands_odd = [0, 0.2, 0.8, 1]
desired_odd = [1, 1]
weights_odd = [1, 1]
b2 = signal.remez(filter_order_odd, bands_odd, desired_odd, weights_odd, fs=2, type='hilbert')

print(b2)
plt.subplot(2, 1, 2)
plt.stem(b2)
plt.title('Type ' + str(firtype(b2)))

#plt.tight_layout()

```

#### FIR digitalFilter对象的类型
