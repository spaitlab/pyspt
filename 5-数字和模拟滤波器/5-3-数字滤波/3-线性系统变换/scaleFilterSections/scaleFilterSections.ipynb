````python
# 信号处理仿真与应用 - 数字和模拟滤波器 - 线性系统变换

## MATLAB函数描述：scaleFilterSections 

函数来源：[MATLAB envelope](https://ww2.mathworks.cn/help/signal/ref/scalefiltersections.html)

### 语法

ctfg = scaleFilterSections(ctf,g)

### 说明

ctfg = scaleFilterSections(ctf,g) 根据 g 中指定的比例值缩放级联传递函数矩阵 ctf 的部分。

### 输入参数

ctf - 分子滤波器系数级联
矩阵
以矩阵形式指定的分子滤波器系数级联。ctf 的行数等于 L，即级联中滤波器部分的数量。ctf 的列数等于 N + 1，其中 N 是分段顺序。
数据类型：single | double | int8 | int16 | int32 | int64 | uint8 | uint16 | uint32 | uint64
支持复数： 支持

g - 比例值
标量 | 矢量
缩放值，指定为标量或包含 L + 1 个元素的向量，其中 L 是级联中滤波器分区的数量。
如果 g 是标量，scaleFilterSections 会将该值统一应用到所有级联滤波器部分。
如果 g 是矢量，则 scaleFilterSections 会将前 L 个标度值分别应用到相应的滤波器分段，并将最后一个标度值统一应用到所有滤波器分段。
数据类型：single | double | int8 | int16 | int32 | int64 | uint8 | uint16 | uint32 | uint64


### 输出参量

ctfg - 按比例计算的分母滤波器系数
矩阵
缩放分母滤波器系数，以矩阵形式返回。ctfg 的大小与 ctf 相同。



## Python函数描述：scaleFilterSections 

函数来源：自定义

### 包络函数定义：


### 参数
- `signal`: 输入信号，可以是任何数组形式的数据，通常是一维的，代表时间序列信号。

### 返回值
- `upper_envelope`: 信号的上包络线，即幅度包络线。这是通过取解析信号的幅度得到的。
- `lower_envelope`: 信号的下包络线。对于对称信号，这将是上包络线的负值。对于非对称信号，这可能并不代表一个真正的下包络线。

### 注意事项
- 该函数并不适用于非对称信号的真正上下包络线的计算。对于非对称信号，需要更高级的技术，如经验模态分解（Empirical Mode Decomposition, EMD）来获取真正的上下包络线。

### 函数工作原理
1. 使用 `hilbert` 函数对输入信号进行希尔伯特变换，得到解析信号。
2. 计算解析信号的幅度，即为信号的幅度包络线。
3. 将幅度包络线赋值给 `upper_envelope` 作为上包络线。
4. 将幅度包络线的负值赋值给 `lower_envelope`。需要注意的是，对于非对称信号，这个下包络线可能并不准确。

### 使用场景
这个函数可以用于分析信号的能量分布，特别是在信号处理、语音分析和振动分析等领域。例如，在处理音频信号时，可以通过计算包络线来检测声音的强度变化；在处理振动信号时，可以通过包络线来识别机械故障的早期迹象。

### 改进建议
- 对于非对称信号，可以考虑实现更高级的算法，如EMD，以提供更准确的包络线计算。
- 可以添加对输入信号的类型和形状的检查，确保函数在不同类型的信号上都能正常工作。
- 可以提供额外的参数，允许用户选择不同的窗函数或滤波器来改善包络线的平滑性和准确性。



## Prompt 1 ： 生成 Python envelope 函数

参考下面MATLAB代码的envelope函数
```
% 生成高斯调制二次 chirp。指定采样率为 2 kHz，信号持续时间为 2 秒。
t = 0:1/2000:2-1/2000;
q = chirp(t-2,4,1/2,6,'quadratic',100,'convex').*exp(-4*(t-1).^2);
plot(t,q)
[up,lo] = envelope(q);
hold on
plot(t,up,t,lo,'linewidth',1.5)
legend('q','up','lo')
hold off
```

和我们采用Python语言实现的解析信号计算啁啾信号的上下包络，
```
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import hilbert

# 生成时间序列 t，采样率为2kHz，持续时间为2秒
t = np.linspace(0, 2, 4000, endpoint=False)

# 生成高斯调制的二次啁啾信号
q = signal.chirp(t, f0=6, f1=4, t1=0.5, method='quadratic', phi=100, vertex_zero=True) * np.exp(-4*(t-1)**2)

# 使用hilbert函数计算解析信号
analytic_signal = hilbert(q)
amplitude_envelope = np.abs(analytic_signal)

# 计算上下包络
up = amplitude_envelope
lo = -amplitude_envelope

# 绘制原始信号和上下包络
plt.plot(t, q, label='Original Signal', color='blue')
plt.plot(t, up, label='Upper Envelope', color='green', linewidth=1.5)
plt.plot(t, lo, label='Lower Envelope', color='red', linewidth=1.5)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Gaussian Modulated Quadratic Chirp Signal and Its Envelopes')
plt.legend()
plt.show()
```
请根据使用解析信号计算啁啾信号的上下包络程序，完成Python语言envelope函数的编写，便于类似应用调用。

### LLM 生成 ： Python envelope 函数

这是包络函数定义
def envelope(signal):
    """
    Compute the envelope of a signal using the Hilbert transform.

    Parameters:
    signal (array_like): The input signal for which the envelope is to be computed.

    Returns:
    upper_envelope (array_like): The upper envelope of the signal (same as amplitude envelope).
    lower_envelope (array_like): For symmetric signals, this will be the negative of the upper envelope.
    For asymmetric signals, this might not represent a true lower envelope.

    Note: This function does not provide true upper and lower envelopes for asymmetric signals.
    For that, more advanced techniques like Empirical Mode Decomposition (EMD) are required.
    """
    analytic_signal = hilbert(signal)
    amplitude_envelope = np.abs(analytic_signal)
    upper_envelope = amplitude_envelope
    lower_envelope = -amplitude_envelope  # This is not a true lower envelope for asymmetric signals.

    return upper_envelope, lower_envelope


## Prompt 2 ： 应用生成的 Python envelope 函数做MATLAB示例对应

这是原始信号q
```
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import hilbert

# 生成时间序列 t，采样率为2kHz，持续时间为2秒
t = np.linspace(0, 2, 4000, endpoint=False)

# 生成高斯调制的二次啁啾信号
q = signal.chirp(t, f0=6, f1=4, t1=0.5, method='quadratic', phi=100, vertex_zero=True) * np.exp(-4*(t-1)**2)
```

这是包络函数定义
def envelope(signal):
    """
    Compute the envelope of a signal using the Hilbert transform.

    Parameters:
    signal (array_like): The input signal for which the envelope is to be computed.

    Returns:
    upper_envelope (array_like): The upper envelope of the signal (same as amplitude envelope).
    lower_envelope (array_like): For symmetric signals, this will be the negative of the upper envelope.
    For asymmetric signals, this might not represent a true lower envelope.

    Note: This function does not provide true upper and lower envelopes for asymmetric signals.
    For that, more advanced techniques like Empirical Mode Decomposition (EMD) are required.
    """
    analytic_signal = hilbert(signal)
    amplitude_envelope = np.abs(analytic_signal)
    upper_envelope = amplitude_envelope
    lower_envelope = -amplitude_envelope  # This is not a true lower envelope for asymmetric signals.

    return upper_envelope, lower_envelope

请给出应用包络函数envelope对原始信号q求包络线，并绘制图形，给出Python程序。

### LLM 生成 ： 
```
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from scipy.signal import hilbert

# 生成时间序列 t，采样率为2kHz，持续时间为2秒
t = np.linspace(0, 2, 4000, endpoint=False)

# 生成高斯调制的二次啁啾信号
q = signal.chirp(t, f0=6, f1=4, t1=0.5, method='quadratic', phi=100, vertex_zero=True) * np.exp(-4*(t-1)**2)

def envelope(signal):  
    """  
    Compute the envelope of a signal using the Hilbert transform.  

    Parameters:  
    signal (array_like): The input signal for which the envelope is to be computed.  

    Returns:  
    upper_envelope (array_like): The upper envelope of the signal (same as amplitude envelope).  
    lower_envelope (array_like): For symmetric signals, this will be the negative of the upper envelope.  
                                For asymmetric signals, this might not represent a true lower envelope.  

    Note: This function does not provide true upper and lower envelopes for asymmetric signals.  
        For that, more advanced techniques like Empirical Mode Decomposition (EMD) are required.  
    """  
    analytic_signal = hilbert(signal)  
    amplitude_envelope = np.abs(analytic_signal)  
    upper_envelope = amplitude_envelope  
    lower_envelope = -amplitude_envelope  # This is not a true lower envelope for asymmetric signals.  
    
    return upper_envelope, lower_envelope  

# 计算包络线
upper_envelope, lower_envelope = envelope(q)

#绘制原始信号和包络线
plt.plot(t, q, label='Original Signal', color='blue')
plt.plot(t, upper_envelope, label='Upper Envelope', color='green', linewidth=1.5)
plt.plot(t, lower_envelope, label='Lower Envelope', color='red', linewidth=1.5)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Gaussian Modulated Quadratic Chirp Signal and Its Envelopes')
plt.legend()
plt.show()
```

## Prompt 3 ： 应用生成的 Python envelope.py 函数做MATLAB示例对应

参考下面Matlab代码，给出Python实现代码：

注意：
(1)Matlab和Python对chirp函数的名称不一样，参数定义不一样，需要换算。请先检索：Matlab网站chirp函数，和Python Scipy网站signal.chirp函数，对比参数，给出换算方式。
(2)线条采用不同颜色方便区分。

**Matlab代码**
```
% 生成高斯调制二次 chirp。指定采样率为 2 kHz，信号持续时间为 2 秒。
t = 0:1/2000:2-1/2000;
q = chirp(t-2,4,1/2,6,'quadratic',100,'convex').*exp(-4*(t-1).^2);
plot(t,q)
% 使用解析信号计算啁啾信号的上下包络。
[up,lo] = envelope(q);
hold on
plot(t,up,t,lo,'linewidth',1.5)
legend('q','up','lo')
hold off
% 由于均值非零，因此，信号不对称。
% 使用不带输出参量的 envelope 以采样编号为 x 轴绘制信号和包络。
envelope(q)
```

这是Python envelope.py
```
"""
This file contains code used in "GPT-PySPT", by Zhiguo Zhou, available from SpaitLab
Copyright 2024 Zhiguo Zhou
License: GNU GPLv3 http://www.gnu.org/licenses/gpl.html
"""
import numpy as np
from scipy.signal import hilbert

def envelope(signal):  
    """  
    Compute the envelope of a signal using the Hilbert transform.  

    Parameters:  
    signal (array_like): The input signal for which the envelope is to be computed.  

    Returns:  
    upper_envelope (array_like): The upper envelope of the signal (same as amplitude envelope).  
    lower_envelope (array_like): For symmetric signals, this will be the negative of the upper envelope.  
                                For asymmetric signals, this might not represent a true lower envelope.  

    Note: This function does not provide true upper and lower envelopes for asymmetric signals.  
        For that, more advanced techniques like Empirical Mode Decomposition (EMD) are required.  
    """  
    analytic_signal = hilbert(signal)  
    amplitude_envelope = np.abs(analytic_signal)  
    upper_envelope = amplitude_envelope  
    lower_envelope = -amplitude_envelope  # This is not a true lower envelope for asymmetric signals.  
  
    return upper_envelope, lower_envelope 
```


### LLM 生成 ： 

下面这是调用程序
```
from envelope import envelope
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 生成时间序列 t，采样率为2kHz，持续时间为2秒
t = np.linspace(0, 2, 4000, endpoint=False)

# 生成高斯调制的二次啁啾信号
q = signal.chirp(t, f0=6, f1=4, t1=0.5, method='quadratic', phi=100, vertex_zero=True) * np.exp(-4*(t-1)**2)

# 计算包络线
upper_envelope, lower_envelope = envelope(q)

# 绘制原始信号和包络线
plt.plot(t, q, label='Original Signal', color='blue')
plt.plot(t, upper_envelope, label='Upper Envelope', color='green', linewidth=1.5)
plt.plot(t, lower_envelope, label='Lower Envelope', color='red', linewidth=1.5)

plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Gaussian Modulated Quadratic Chirp Signal and Its Envelopes')
plt.legend()
plt.show()
```
````
