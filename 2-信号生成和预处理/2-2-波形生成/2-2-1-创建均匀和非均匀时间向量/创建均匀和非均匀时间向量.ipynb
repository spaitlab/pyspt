### - <font color=DarkOrChid size=3>示例2：创建非均匀时间向量</font>

```python
Ffast = 100
Tf = 1/Ffast
Nslow = 5
#tdisc = [linspace(0,1,Nslow) 1+Tf:Tf:2-Tf linspace(2,3,Nslow)]
tdisc = np.linspace(0,1,Nslow)
#Tdisc = np.array(tdisc)
Nfast = (2-1)/Tf
#print(Nfast)
tdisc1 = np.linspace(1,2,99)
tdisc = np.append(tdisc,tdisc1)
tdisc2 = np.linspace(2,3,Nslow)
tdisc = np.append(tdisc,tdisc2)
```

```python
plt.plot(tdisc,gpl2['gpl2'].flatten(),'o',ms = 5,mfc='w',mew=1,mec='b')
plt.plot(Tcont,gpl1['gpl1'])
plt.title('Triangular_pulse Wave')
plt.xlabel('Time (sec)')
plt.ylabel('Amplitude')
```
