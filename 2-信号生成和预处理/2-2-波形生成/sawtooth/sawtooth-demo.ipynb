## sawtooth

#### 锯齿波

**和MATLAB波形一致**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth

T = 10*(1/50)
fs = 1000
t = np.arange(0, T, 1/fs)

x = sawtooth(2*np.pi*50*t)

plt.plot(t, x)
plt.grid(True)
plt.show()
```

#### 三角波

**和MATLAB波形一致**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth

T = 10*(1/50)
fs = 1000
t = np.arange(0, T, 1/fs)

x = sawtooth(2*np.pi*50*t, width=0.5)

plt.plot(t, x)
plt.grid(True)
plt.show()
```
